```python
import numpy as np
import random as rd


def sample_entropy(row_vec_p, emb_dim, tolr_intr):
    pt_add, nb_add, na_add, m_add, axis_val_a, axis_val_b =  (1, 1, 1, 1, 1, 1)
    nr_data_tot, a_arr, b_arr = len(row_vec_p), 0.0, 0.0
    m_i = np.array([row_vec_p [i : i + emb_dim] for i in range(nr_data_tot -  emb_dim)])
    m_j = np.array([row_vec_p [i : i + emb_dim] for i in range(nr_data_tot - emb_dim + pt_add)])
    b_arr = np.sum([np.sum(np.abs(i - m_j).max(axis=axis_val_b) <= tolr_intr) - nb_add for i in m_i])
    emb_dim += m_add
    m_x = np.array([row_vec_p [i : i + emb_dim] for i in range(nr_data_tot - emb_dim + m_add)])
    a_arr = np.array([np.sum(np.abs(m_i - m_x).max(axis=axis_val_a) <= tolr_intr) - 1 for m_i in m_x])
    return -np.log(a_arr/b_arr)
```

```python
rd_ls = np.linspace(-1000,1000,102, 1).tolist()
rd.shuffle(rd_ls)
arr_z = sample_entropy(rd_ls, 2, 400)

arr_z
```
